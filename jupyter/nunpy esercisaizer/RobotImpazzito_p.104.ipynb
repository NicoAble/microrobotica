```python
%%timeit
import random
posizione=0
secondi = 60 * 60 * 24 * 5
for i in range(secondi):
    posizione += (2 * random.randint(0,1))-1
print(f"Spostamento accumulato rispetto alla posizione centrale: {posizione} cm")
```

```python
%%timeit
import numpy as np

secondi = 60 * 60 * 24 * 5
movimenti = np.random.choice([-1,+1], secondi)
posizione = np.sum(movimenti)
print(f"Spostamento accumulato rispetto alla posizione centrale: {posizione} cm")
```
