```python
import string #per segni di punteggiatura
```

```python
def leggi_file(nome_file):
    """La funzione legge i file"""

    file = open(nome_file, "r")

    #per leggere un file e salvare le righe in una lista
    lista_righe = file.readlines()

    #print(lista_righe)

    file.close()
    lista_numeri=[]
    diz_parole={"parole":[], "numero":[]}

    for riga in lista_righe[1:]:
        riga_senza_linefeed = riga[:-1]
        lista_parole = riga_senza_linefeed.split(" ") #è un metodo, lo spazio è il separatore
        
        for i, s in enumerate(lista_parole):
            lista_parole[i] = s.translate(str.maketrans('', '', string.punctuation))
        
        for parola in lista_parole: #lista di tutte le parole individuate
            #print(parola)
            if parola!="":
                if len(diz_parole["parole"])==0:
                    diz_parole["parole"].append(parola)
                    lista_numeri.append(1)
                else:
                    j=False #variabile che controlla che non ci siano parole nuove
                    for i,p in enumerate(diz_parole["parole"]): 
                        if parola==p: 
                            j=True
                            lista_numeri[i]+=1 #quante volte compare la parola
                    if j==False: 
                        diz_parole["parole"].append(parola)
                        lista_numeri.append(1)
    diz_parole["numero"].append(lista_numeri)                
              
    return diz_parole
```

```python
diz= leggi_file("./sei_personaggi_in_cerca_dautore_Luigi_Pirandello.txt")
lista_numeri=diz["numero"]
lista_numeri=lista_numeri[0]
#print(lista_numeri)
lista_parole=diz["parole"]
for i, p in enumerate(lista_parole):
    print(p,"\t",lista_numeri[i])
```
